Outliers

```python
import pandas as pd
def find_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    l_value=q1-(1.5*IQR)
    h_value=q3+(1.5*IQR)
    print(l_value)
    print(h_value)
    ds=ds.loc[(ds[col]<l_value) | (ds[col]>h_value)]
    return ds
cars=pd.read_csv('auto-mpg.csv')
find_outliers(cars,'mpg')
```

```python
find_outliers(cars,'acceleration')
```

```python
import pandas as pd
def remove_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    l_value=q1-(1.5*IQR)
    h_value=q3+(1.5*IQR)
    print(l_value)
    print(h_value)
    ds=ds.loc[(ds[col]>=l_value) & (ds[col]<=h_value)]
    return ds
cars=pd.read_csv('auto-mpg.csv')
remove_outliers(cars,'mpg')
```

```python
remove_outliers(cars,'acceleration')
```

```python
import pandas as pd
df=pd.DataFrame({'Dept':['Sales','Sales','HR','HR','IT','IT'],
                 'Salary':[50000,60000,40000,45000,70000,75000],
                 'Bonus':[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Dept').agg(['mean','min','max'])
result
```

```python
import pandas as pd
df=pd.DataFrame({'Dept':['Sales','Sales','HR','HR','IT','IT'],
                 'Salary':[50000,60000,40000,45000,70000,75000],
                 'Bonus':[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Dept').agg({'Salary':['min','max'],
                               'Bonus':['sum']})
result
```
